```python
import os
from astropy.io import fits

# Caminho base
data = os.path.expanduser('/media/fernando/KINGSTON/Doutorado/2025/20250309/')
pastas = ['sparc4acs1', 'sparc4acs2', 'sparc4acs3', 'sparc4acs4']

# Critérios corretos para identificação das imagens de HD298383-L4
ra_hd298383 = '09:22:29.77'
wp_hd298383 = 'L4'
object_errado = 'Vela1-95'
object_correto = 'HD298383'

for pasta in pastas:
    caminho_pasta = os.path.join(data, pasta)
    for nome_arquivo in os.listdir(caminho_pasta):
        if nome_arquivo.endswith('.fits'):
            caminho_arquivo = os.path.join(caminho_pasta, nome_arquivo)
            with fits.open(caminho_arquivo, mode='update') as hdul:
                header = hdul[0].header
                # Verifica se é uma imagem problemática
                if (
                    header.get('WPSEL') == wp_hd298383 and
                    header.get('RA') == ra_hd298383 and
                    header.get('OBJECT') == object_errado
                ):
                    # Corrige OBJECT
                    header['OBJECT'] = object_correto
                    print(f"[OK] Corrigido OBJECT em: {nome_arquivo}")

                    # Corrige FILENAME (se existir e contiver "vela")
                    if 'FILENAME' in header and 'vela' in header['FILENAME'].lower():
                        antigo = header['FILENAME']
                        novo = antigo.lower().replace('vela', 'hd29')
                        header['FILENAME'] = novo
                        print(f"     -> FILENAME alterado: {antigo} → {novo}")
                else:
                    print(f"[--] Sem alterações: {nome_arquivo}")
```
